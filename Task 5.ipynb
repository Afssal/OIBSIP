```python
#import the packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
#read csv file
df=pd.read_csv("D:\\gan\\internship\\task_4_sales\\advertising.csv")
```

```python
df
```

```python
df.columns
```

```python
li=['TV', 'Radio', 'Newspaper']
```

```python
#predictor
X=df[li]
```

```python
#target variable
Y=df['Sales']
```

```python
#spliting dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_test
```

```python
X_train
```

```python
Y_train
```

```python
#model building
model = LinearRegression()
```

```python
#model training
model.fit(X_train,Y_train)
```

```python
#model predicting
predictor=model.predict(X_test)
```

```python
#mean absolute error checking
mean_absolute_error(Y_test, predictor)
```
